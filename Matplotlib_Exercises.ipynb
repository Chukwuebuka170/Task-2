## 1. Read total profit of all months and show it using a lne plot.

```python
# Importing the libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Loading the Dataset
df = pd.read_csv('company_sales_data.csv')
df
```

```python
# Exercise 1: Line plot of total profit

plt.figure(figsize=(10,5))
plt.plot(df['month_number'], df['total_profit'], marker='o', color='b', linestyle='-')
plt.title("Total Profit of All Months")
plt.xlabel("Month")
plt.ylabel("Profit")
plt.grid(True)
plt.show()
```

# 2. Read data for bathing soap and face wash of all months and display it using subplot

```python
# Showing Subplots of Bathing Soap and Facewash (side by side)

fig, ax = plt.subplots(1, 2, figsize=(12, 5))

# Subplot 1: Bathing Soap
ax[0].plot(df['month_number'], df['bathingsoap'], marker='o', color='r', linestyle='-')
ax[0].set_title("Bathing Soap Sales Data")
ax[0].set_xlabel("Month Number")
ax[0].set_ylabel("Units Sold")
ax[0].grid(True)

# Subplot 2: Facewash
ax[1].plot(df['month_number'], df['facewash'], marker='o', color='g', linestyle='-')
ax[1].set_title("Facewash Sales Data")
ax[1].set_xlabel("Month Number")
ax[1].set_ylabel("Units Sold")
ax[1].grid(True)
```
